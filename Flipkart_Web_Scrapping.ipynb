# Flipkart review Scrapping

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

Sample url:
https://www.flipkart.com/asus-rog-5s-storm-white-256-gb/product-reviews/itm360bb6ee4b326?pid=MOBG6Y6HZZGFZZAU&lid=LSTMOBG6Y6HZZGFZZAUHUZGYX&marketplace=FLIPKART

```python
url=input("Please paste the url for which you want to extract reviews:")
```

```python
pages=requests.get(url)
pages.text
#to have html pages in streuctured format
soup=BeautifulSoup(pages.text,'html.parser')
print(soup.prettify())
```

```python
#extract review
rev=soup.find_all('p',class_='_2-N8zT')
review=[]
for i in rev:
    i=i.get_text().replace('\n','')
    i=i.strip('')
    review.append(i)
review
```

```python
#extract rating
rt=soup.find_all('div',class_=['_3LWZlK _1BLPMq','_3LWZlK _1rdVr6 _1BLPMq','_3LWZlK _32lA32 _1BLPMq'])
ratings=[]
for i in rt:
    i=i.get_text().replace('\n','')
    i=i.strip()
    ratings.append(i)
ratings
```

```python
#Extract Name and time
ti=soup.find_all('p',class_='_2sc7ZR')
Name_Time=[]
for i in ti:
    i=i.get_text().replace('\n','')
    i=i.strip()
    Name_Time.append(i)
Name_Time
```

```python
#Seperate name and time
name=[]
Time=[]
for i in Name_Time:
    if Name_Time.index(i) == 0 or Name_Time.index(i)%2==0:
        name.append(i)
    else:
        Time.append(i)
print(name)
print(Time)
```

```python
#To create Data_Frame
df=pd.DataFrame()
df["Reviews"]=review
df["Ratings"]=ratings
df["Name"]=name
df["Time"]=Time
df
```
